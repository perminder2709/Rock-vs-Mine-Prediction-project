```python
#Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
```

```python
synthetic_data = np.random.multivariate_normal(
    mean=[0, 0, 0],
    cov=[[1, 0.8, 0.5], [0.8, 1, 0.3], [0.5, 0.3, 1]],
    size=300
)
df = pd.DataFrame(synthetic_data, columns=['Feature 1', 'Feature 2', 'Feature 3'])
correlation_matrix = df.corr()
plt.subplot(3, 2, 5)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap: Correlation Matrix')
```

```python
#. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
np.random.seed(30)
data_3d = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df_3d = pd.DataFrame(data_3d)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_3d['X'], df_3d['Y'], df_3d['Z'], c='blue', alpha=0.6)
ax.set_title('3D Scatter Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

```python
# Heatmap for Sales Data
```

```python
np.random.seed(20)
data_sales = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data_sales)
heatmap_data = pd.pivot_table(df_sales, values='Sales', index='Day', columns='Month', aggfunc='mean')

plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, fmt='.0f', cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap: Sales Data')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
```

```python
# Using the given x and y data, generate a 3D surface plot to visualize the function 
#3D Surface Plot for the Function z = sin(sqrt(x^2 + y^2))
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
ax.set_title('3D Surface Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=5)
plt.show()
```

```python
# Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.
```

```python
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.6, c=np.random.rand(5), cmap='viridis', edgecolors='black')
for i, country in enumerate(df['Country']):
    plt.text(df['GDP'][i], df['Population'][i] + 20, country, fontsize=9, ha='center')
plt.title('Bubble Chart: Country Population vs GDP')
plt.xlabel('GDP')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
#.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import dodge, linear_cmap
from bokeh.layouts import column
import numpy as np
import pandas as pd

output_notebook()
```

```python
x = np.linspace(0, 10, 500)
y = np.sin(x)

p1 = figure(title="Sine Wave", x_axis_label='X', y_axis_label='Sin(X)', width=600, height=400)
p1.line(x, y, legend_label='Sine Wave', line_width=2, color="blue")
show(p1, notebook_handle=True)
```

```python
#Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
```

```python
np.random.seed(42)
x_scatter = np.random.rand(50) * 10
y_scatter = np.random.rand(50) * 10
sizes = np.random.randint(10, 50, 50)
colors = np.random.randint(100, 255, 50)

source_scatter = ColumnDataSource(data=dict(x=x_scatter, y=y_scatter, size=sizes, color=colors))

p2 = figure(title="Bokeh Scatter Plot", x_axis_label='X', y_axis_label='Y', width=600, height=400)
p2.scatter(x='x', y='y', size='size', color=linear_cmap('color', 'Viridis256', 100, 255), alpha=0.6, source=source_scatter)
show(p2, notebook_handle=True)
```
