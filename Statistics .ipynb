````python
import random
import numpy as np
from scipy import stats

# Generate a list of 100 integers between 90 and 130
np.random.seed(42)  # For reproducibility
int_list = [random.randint(90, 130) for _ in range(100)]

# (i) Mean and Median
def calculate_mean(data):
    return np.mean(data)

def calculate_median(data):
    return np.median(data)

mean_val = calculate_mean(int_list)
median_val = calculate_median(int_list)

# (ii) Mode
def calculate_mode(data):
    try:
        # For newer versions of scipy
        mode_val = stats.mode(data, keepdims=True)
    except TypeError:
        # For older versions of scipy
        mode_val = stats.mode(data)
    return mode_val.mode[0]

mode_val = calculate_mode(int_list)

# (iii) Weighted Mean
def calculate_weighted_mean(values, weights):
    return np.average(values, weights=weights)

# Fix: Generate weights with the correct length
weights = [random.random() for _ in range(len(int_list))]
weighted_mean_val = calculate_weighted_mean(int_list, weights)

# (iv) Geometric Mean
def calculate_geometric_mean(data):
    return stats.gmean(data)

geometric_mean_val = calculate_geometric_mean(int_list)

# (v) Harmonic Mean
def calculate_harmonic_mean(data):
    return stats.hmean(data)

harmonic_mean_val = calculate_harmonic_mean(int_list)

# (vi) Midrange
def calculate_midrange(data):
    return (max(data) + min(data)) / 2

midrange_val = calculate_midrange(int_list)

# (vii) Trimmed Mean
def calculate_trimmed_mean(data, proportion):
    return stats.trim_mean(data, proportion)

trimmed_mean_val = calculate_trimmed_mean(int_list, 0.1)

# Display all results
print(f"Dataset: {int_list[:10]}... (showing first 10 values)")
print(f"Number of values: {len(int_list)}")
print(f"Range: {min(int_list)} to {max(int_list)}")
print("\nMeasures of Central Tendency:")
print(f"1. Arithmetic Mean: {mean_val:.4f}")
print(f"2. Median: {median_val:.4f}")
print(f"3. Mode: {mode_val}")
print(f"4. Weighted Mean: {weighted_mean_val:.4f}")
print(f"5. Geometric Mean: {geometric_mean_val:.4f}")
print(f"6. Harmonic Mean: {harmonic_mean_val:.4f}")
print(f"7. Midrange: {midrange_val:.4f}")
print(f"8. 10% Trimmed Mean: {trimmed_mean_val:.4f}")

# Additional statistics
print("\nAdditional Statistics:")
print(f"Standard Deviation: {np.std(int_list):.4f}")
print(f"Variance: {np.var(int_list):.4f}")
````

````python
# Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.
class DiscreteRandomVariable:
    def __init__(self, outcomes, probabilities):
        self.outcomes = outcomes
        self.probabilities = probabilities

    def expected_value(self):
        return sum(o * p for o, p in zip(self.outcomes, self.probabilities))

    def variance(self):
        mean = self.expected_value()
        return sum(p * (o - mean) ** 2 for o, p in zip(self.outcomes, self.probabilities))

# Example usage:
outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [1/6] * 6
drv = DiscreteRandomVariable(outcomes, probabilities)
expected = drv.expected_value()
variance = drv.variance()
````

````python
#Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance
from scipy.stats import poisson, binom

def generate_samples(distribution, params, size=1000):
    if distribution == "binomial":
        samples = binom.rvs(n=params['n'], p=params['p'], size=size)
    elif distribution == "poisson":
        samples = poisson.rvs(mu=params['mu'], size=size)
    else:
        raise ValueError("Unsupported distribution")
    return np.mean(samples), np.var(samples)

mean, var = generate_samples("poisson", {'mu': 5}, 1000)
````

````python
#Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute the mean, variance, and standard deviation of the samples.
def gaussian_samples(mean, variance, size=1000):
    std_dev = np.sqrt(variance)
    samples = np.random.normal(mean, std_dev, size)
    return np.mean(samples), np.var(samples), np.std(samples)

mean, variance, std_dev = gaussian_samples(10, 4, 1000)
````

````python
#Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.
import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution.

    Parameters:
        x (float): The value of the random variable.
        mean (float): The mean of the normal distribution.
        std_dev (float): The standard deviation of the normal distribution.

    Returns:
        float: The PDF value at x.
    """
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = math.exp(-((x - mean) ** 2) / (2 * (std_dev ** 2)))
    return coefficient * exponent

# Example usage
x_value = 2.0
mean = 0.0
std_dev = 1.0
pdf_value = normal_pdf(x_value, mean, std_dev)
print(f"The PDF value at x = {x_value} is {pdf_value}")
````

````python
#) Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.
import math

def exponential_cdf(x, lambda_):
    """
    Calculate the cumulative distribution function (CDF) of an exponential distribution.

    Parameters:
        x (float): The value at which to calculate the CDF.
        lambda_ (float): The rate parameter (λ) of the exponential distribution (must be > 0).

    Returns:
        float: The CDF value at x.
    """
    if x < 0:
        return 0.0
    return 1 - math.exp(-lambda_ * x)

# Example usage
x_value = 2.0
lambda_param = 1.5  # Rate parameter (λ)
cdf_value = exponential_cdf(x_value, lambda_param)
print(f"The CDF value at x = {x_value} for λ = {lambda_param} is {cdf_value}")
````

````python
#) Write a Python function to calculate the probability mass function (PMF) of Poisson distribution
import math

def poisson_pmf(k, lambda_):
    """
    Calculate the probability mass function (PMF) of a Poisson distribution.

    Parameters:
        k (int): The number of occurrences (non-negative integer).
        lambda_ (float): The mean rate (λ) of the Poisson distribution (must be > 0).

    Returns:
        float: The PMF value for k.
    """
    if k < 0 or not isinstance(k, int):
        raise ValueError("k must be a non-negative integer.")
    return (math.exp(-lambda_) * (lambda_ ** k)) / math.factorial(k)

# Example usage
k_value = 3  # Number of occurrences
lambda_param = 2.5  # Mean rate (λ)
pmf_value = poisson_pmf(k_value, lambda_param)
print(f"The PMF value for k = {k_value} and λ = {lambda_param} is {pmf_value}")
````

````python
#M A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors
who make a purchase). They collect data from the old and new layouts to compare.


To generate the data use the following command:

```python

import numpy as np

# 50 purchases out of 1000 visitors

old_layout = np.array([1] * 50 + [0] * 950)

# 70 purchases out of 1000 visitors  

new_layout = np.array([1] * 70 + [0] * 930)

  ```

Apply z-test to find which layout is successful.

````

````python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Generate the data
old_layout = np.array([1] * 50 + [0] * 950)  # 50 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)  # 70 purchases out of 1000 visitors

# Calculate the number of successes and total observations for both groups
successes = [np.sum(old_layout), np.sum(new_layout)]
n_obs = [len(old_layout), len(new_layout)]

# Perform a two-proportion z-test
z_stat, p_value = proportions_ztest(successes, n_obs)

# Interpret the results
alpha = 0.05  # Significance level
print(f"Z-Statistic: {z_stat:.3f}")
print(f"P-Value: {p_value:.3f}")

if p_value < alpha:
    print("Reject the null hypothesis: The new layout is statistically more successful.")
else:
    print("Fail to reject the null hypothesis: No significant difference between layouts.")
````

````python
# A tutoring service claims that its program improves students' exam scores. A sample of students who
participated in the program was taken, and their scores before and after the program were recorded.


Use the below code to generate samples of respective arrays of marks:

```python

before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])

after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

```

Use z-test to find if the claims made by tutor are true or false.
````

````python
import numpy as np
from scipy.stats import norm

# Data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Differences in scores
differences = after_program - before_program

# Calculate the mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)

# Sample size
n = len(differences)

# Z-test statistic
z_stat = mean_diff / (std_diff / np.sqrt(n))

# P-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # Two-tailed test

# Significance level
alpha = 0.05

# Output results
print(f"Mean Difference: {mean_diff:.2f}")
print(f"Z-Statistic: {z_stat:.3f}")
print(f"P-Value: {p_value:.3f}")

if p_value < alpha:
    print("Reject the null hypothesis: The tutoring program significantly improved exam scores.")
else:
    print("Fail to reject the null hypothesis: No significant improvement in exam scores.")
````

````python
#A company is testing two different website layouts to see which one leads to higher click-through rates.
Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of
freedom, and p-value.


Use the following data:

```python

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]

layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]
````

````python
from scipy.stats import ttest_ind

def ab_test(layout_a_clicks, layout_b_clicks, alpha=0.05):
    """
    Perform an A/B test analysis between two groups.
    
    Parameters:
        layout_a_clicks (list): Click-through rates for layout A.
        layout_b_clicks (list): Click-through rates for layout B.
        alpha (float): Significance level for the test (default is 0.05).
        
    Returns:
        None
    """
    # Perform a two-sample t-test
    t_stat, p_value = ttest_ind(layout_a_clicks, layout_b_clicks, equal_var=False)
    
    # Calculate degrees of freedom
    n1, n2 = len(layout_a_clicks), len(layout_b_clicks)
    s1, s2 = np.var(layout_a_clicks, ddof=1), np.var(layout_b_clicks, ddof=1)
    df = ((s1 / n1 + s2 / n2) ** 2) / (((s1 / n1) ** 2) / (n1 - 1) + ((s2 / n2) ** 2) / (n2 - 1))
    
    # Output results
    print(f"T-Statistic: {t_stat:.3f}")
    print(f"Degrees of Freedom: {df:.2f}")
    print(f"P-Value: {p_value:.3f}")
    
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference between the layouts.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between the layouts.")

# Data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform A/B test
ab_test(layout_a_clicks, layout_b_clicks)
````

````python
from scipy import stats

def analyze_intervention(pre_scores, post_scores):
    # Perform a paired t-test
    t_stat, p_value = stats.ttest_rel(post_scores, pre_scores)
    
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    
    # Interpret results
    if p_value < 0.05:
        print("Result: The intervention had a statistically significant impact on scores.")
    else:
        print("Result: The intervention did NOT have a statistically significant impact on scores.")

# Provided test scores
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Run the analysis
analyze_intervention(pre_intervention_scores, post_intervention_scores)
````

````python
import numpy as np
from scipy import stats

# Generate synthetic salary data for male and female employees
np.random.seed(0)  # For reproducibility

male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

def analyze_gender_salary_gap(male_salaries, female_salaries):
    # Perform an independent two-sample t-test (assuming unequal variances)
    t_stat, p_value = stats.ttest_ind(male_salaries, female_salaries, equal_var=False)
    
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    
    # Interpretation with significance level 0.05
    if p_value < 0.05:
        print("Result: There is a statistically significant difference between male and female salaries.")
    else:
        print("Result: There is NO statistically significant difference between male and female salaries.")

# Run the analysis
analyze_gender_salary_gap(male_salaries, female_salaries)
````
